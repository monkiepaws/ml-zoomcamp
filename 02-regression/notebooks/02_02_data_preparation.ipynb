# Module 2.2 Data preparation

## Setup

### Imports

```python
import numpy as np
import pandas as pd
```

### Load data

```python
# Setup data
df = pd.read_csv('02-regression/notebooks/data.csv')
df.sample(5)
```

## Data preparation

```python
# Make column names all lowercase and replace spaces with underscores
df.columns = df.columns.str.lower().str.replace(' ', '_')
list(df.columns)
```

```python
strings = list(df.dtypes[df.dtypes == 'object'].index)
strings
```

```python
for col in strings:
    df[col] = df[col].str.lower().str.replace(' ', '_')

df.sample(5)
```

## Export dataset

```python
# Save cleaned data for use in next steps
df.to_csv('02-regression/notebooks/data_cleaned.csv', index=False)
```
